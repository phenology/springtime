# PPO

Observations from the [Plant Phenology Ontology (PPO)](https://www.ebi.ac.uk/ols/ontologies/ppo)

---

Uses [rppo] to get data from http://plantphenology.org/

See [dataset documentation](https://github.com/PlantPhenoOntology/ppo/blob/master/documentation/ppo.pdf)


```python
from springtime.datasets import RPPO
from rich import print

# Phenophase: PPO:0002032 = flower presence
dataset = RPPO(
    genus="Syringa", termID="obo:PPO_0002032", limit=1000, years=[1990, 2020]
)
print(dataset)
```

```python
import logging

logging.basicConfig(level=logging.INFO)
raw_df = dataset.raw_load()
```
